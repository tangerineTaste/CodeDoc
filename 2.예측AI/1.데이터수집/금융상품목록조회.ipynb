## 1. 예금 상품 조회

```python
#정기예금 상품 조회
try:
    load_dotenv()
    api_key = os.getenv("FSS_API_KEY")
    topFinGrpNo = "020000"
    pageNo=1
    url = "http://finlife.fss.or.kr/finlifeapi/depositProductsSearch.json"
    params = {
        'auth': api_key,
        'topFinGrpNo': topFinGrpNo,
        'pageNo': pageNo
    }

    response = requests.get(url, params = params)
    response.raise_for_status()
    result = response.json()
    print(result)

except requests.exceptions.RequestException as e:
            print(f"정기예금 상품 조회 오류: {e}")
            raise
```

```python
# 정기예금 상품 리스트 추출(기본상품)
import pandas as pd
baseinfo_list = result['result']['baseList']

# DataFrame 변환
baseinfo_df = pd.DataFrame(baseinfo_list)

# 출력 또는 병합
print("기본 상품 정보:")
print(baseinfo_df.head())

```

```python
# 정기예금 상품 리스트 추출(옵션)
option_list = result['result']['optionList']

# DataFrame 변환
option_df = pd.DataFrame(option_list)

# 출력 또는 병합
print("옵션 정보:")
print(option_df.head())
```
