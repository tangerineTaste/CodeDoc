```python
import matplotlib.pyplot as plt
# 한글 폰트 설정 (필요시)
import matplotlib.pyplot as plt
import matplotlib
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows용 한글 폰트
plt.rcParams['axes.unicode_minus'] = False     # 마이너스(-) 깨짐 방지

# 모델 이름
models = ["RandomForest", "CatBoost", "XGBoost", "LightGBM"]

# 각 성능 지표
accuracy =     [0.758, 0.686, 0.706, 0.798]
recall =       [0.949, 0.938, 0.906, 0.957]   
precision =    [0.922, 0.902, 0.942, 0.935]   
f1_score =     [0.924, 0.903, 0.908, 0.936]   

metrics = [accuracy, recall, precision, f1_score]
labels = ["Accuracy", "Recall", "Precision", "F1-score"]
colors = plt.cm.Paired(range(4))

x = range(len(models))
bar_width = 0.15

plt.figure(figsize=(14, 6))

for i, metric in enumerate(metrics):
    plt.bar([p + i * bar_width for p in x], metric, width=bar_width, label=labels[i], color=colors[i])

plt.xticks([p + 2 * bar_width for p in x], models)
plt.ylim(0, 1.1)
plt.ylabel("Score")
plt.title("금융 상품 매칭 모델 성능 비교 (4가지 지표)")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.savefig('ML성능비교.png')
plt.show()
```
